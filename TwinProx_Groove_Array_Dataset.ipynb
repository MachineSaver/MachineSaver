```python
from dataclasses import dataclass
from typing import List, Optional
import numpy as np
from datetime import datetime
import minimalmodbus
import time
from tqdm import tqdm
import math
import plotly.graph_objects as go
from plotly.subplots import make_subplots
import plotly.io as pio

@dataclass
class WaveformData:
    channel_A: np.ndarray
    channel_B: np.ndarray
    groove_array: np.ndarray

    @property
    def combined_waveforms(self) -> np.ndarray:
        return np.concatenate([self.channel_A, self.channel_B])

    def scale_waveforms(self, scale_factor: float = 100.0) -> None:
        self.channel_A = self.channel_A / scale_factor
        self.channel_B = self.channel_B / scale_factor

@dataclass
class CaptureData:
    timestamp: int
    header: List[int]
    waveform_data: WaveformData
    clip_size: int
    raw_data: Optional[List[int]] = None

    @classmethod
    def from_raw_data(cls, timestamp: int, clip_size: int, data_array: List[int]) -> 'CaptureData':
        # Extract header
        header = data_array[:8]

        # Extract waveforms
        raw_waveforms = data_array[8:3608]  # 8 + 3600
        channel_A = np.array(raw_waveforms[:1800], dtype=float)
        channel_B = np.array(raw_waveforms[1800:], dtype=float)

        # Extract groove array
        groove_array = np.array(data_array[3608:], dtype=int)

        # Create WaveformData instance
        waveform_data = WaveformData(
            channel_A=channel_A,
            channel_B=channel_B,
            groove_array=groove_array
        )

        return cls(
            timestamp=timestamp,
            header=header,
            waveform_data=waveform_data,
            clip_size=clip_size,
            raw_data=data_array
        )

    def process_data(self) -> None:
        """Process the waveform data by scaling and applying groove array masks"""
        # Scale the waveforms
        self.waveform_data.scale_waveforms()

        # Apply groove array masks using np.full for proper gap creation
        for i in range(0, len(self.waveform_data.groove_array), 2):
            if i + 1 >= len(self.waveform_data.groove_array):
                break

            start = self.waveform_data.groove_array[i]
            stop = self.waveform_data.groove_array[i + 1]

            # Skip invalid pairs
            if start >= stop:
                continue

            # Ensure indices are within bounds
            start = max(0, start)
            stop = min(stop, len(self.waveform_data.channel_A))

            if start < stop:  # Additional check to ensure valid slice
                nan_array = np.full(stop - start, np.nan)
                self.waveform_data.channel_A[start:stop] = nan_array
                self.waveform_data.channel_B[start:stop] = nan_array

class TwinProx:
    def __init__(self, port: str, slaveaddress: int):
        self.slave = minimalmodbus.Instrument(port=port, slaveaddress=slaveaddress)
        self.configure_modbus_settings()

    def configure_modbus_settings(self):
        self.slave.serial.baudrate = 115200
        self.slave.serial.bytesize = 8
        self.slave.serial.parity = "N"
        self.slave.serial.stopbits = 1
        self.slave.serial.timeout = 0.10
        self.slave.close_port_after_each_call = True
        self.slave.mode = minimalmodbus.MODE_RTU
        self.slave.clear_buffers_before_each_transaction = True

    def setup_capture(self, modbus_channel_ab: int = 6) -> int:
        self.slave.write_register(32, modbus_channel_ab)
        return self.slave.read_register(32)

    def trigger_capture(self) -> int:
        self.slave.write_register(33, 1)
        return int(str(time.time())[slice(10)])

    def check_capture_engine(self) -> int:
        return self.slave.read_register(34, functioncode=3)

    def get_dataclip_size(self) -> int:
        return self.slave.read_long(36, functioncode=3)

    def extract_data(self, clip_size: int) -> List[int]:
        data_array = []
        total_chunks = math.ceil((clip_size / 122))

        for _ in tqdm(range(total_chunks), desc="Extracting Data", ncols=100):
            chunk = self.slave.read_registers(49, 123, functioncode=3)
            chunk.pop(0)
            data_array.extend(chunk)

        return data_array[:clip_size]

    def capture_data(self) -> CaptureData:
        """Capture and return structured data from the device"""
        timestamp = self.trigger_capture()

        while self.check_capture_engine() != 0:
            pass

        clip_size = self.get_dataclip_size()
        data_array = self.extract_data(clip_size)

        # Create structured data object
        capture_data = CaptureData.from_raw_data(
            timestamp=timestamp,
            clip_size=clip_size,
            data_array=data_array
        )

        # Process the data
        capture_data.process_data()

        return capture_data

def create_marker_sizes(channel: np.ndarray) -> List[int]:
    mask = np.isnan(channel)
    first_dot_after_nan = np.concatenate(([False], np.diff(mask) == -1))
    return [20 if x else 1 for x in first_dot_after_nan]

def set_machine_saver_theme():
    """Set up the Machine Saver plotly theme."""
    pio.templates["draft"] = go.layout.Template(
        layout_annotations=[
            dict(
                name="draft watermark",
                text="Machine Saver Inc.",
                textangle=-30,
                opacity=0.1,
                font=dict(color="white", size=50),
                xref="paper",
                yref="paper",
                x=0.5,
                y=0.5,
                showarrow=True,
            )
        ],
        data_scatter=[
            dict(
                marker=dict(size=2),
            )
        ]
    )
    pio.templates.default = "plotly_dark+draft"

def create_combined_plot(capture_data: CaptureData) -> go.Figure:
    set_machine_saver_theme()

    # Create waveform traces
    trace_A = go.Scatter(
        y=capture_data.waveform_data.channel_A,
        mode='lines+markers',
        name='Channel A'
    )
    trace_B = go.Scatter(
        y=capture_data.waveform_data.channel_B,
        mode='lines+markers',
        name='Channel B'
    )

    # Create orbit plot trace
    marker_sizes = create_marker_sizes(capture_data.waveform_data.channel_A)
    orbit_trace = go.Scatter(
        x=capture_data.waveform_data.channel_A,
        y=capture_data.waveform_data.channel_B,
        mode='lines+markers',
        marker=dict(size=marker_sizes, opacity=0.8)
    )

    # Create and return the figure
    fig = make_subplots(rows=1, cols=2, subplot_titles=('Waveform', 'Orbit Plot'))
    fig.add_trace(trace_A, row=1, col=1)
    fig.add_trace(trace_B, row=1, col=1)
    fig.update_yaxes(title_text='Values', row=1, col=1)
    fig.add_trace(orbit_trace, row=1, col=2)
    fig.update_xaxes(title_text='Channel A', row=1, col=2)
    fig.update_yaxes(title_text='Channel B', row=1, col=2)
    fig.update_layout(
        xaxis2=dict(scaleanchor="y2", scaleratio=1),
        title='Combined Plot',
        xaxis_title='Samples'
    )

    return fig

def main():
    twinprox = TwinProx(port="COM4", slaveaddress=10)

    while True:
        # Capture and process data
        capture_data = twinprox.capture_data()

        # Create and display plot
        fig = create_combined_plot(capture_data)
        fig.show()

if __name__ == "__main__":
    main()
```
